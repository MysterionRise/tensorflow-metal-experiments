```python
import matplotlib.pyplot as plt
import numpy as np


plt.rcdefaults()
fig, ax = plt.subplots()

# Example data
hardware = ('RTX2070 8Gb', 'M1 Max 32-core GPU', 'M2 10-core GPU',  'M1 Max 10-core CPU', 'M2 8-core CPU', '2.4 Ghz 8-core i9', '3.2 Ghz 6-core i7')
y_pos = np.arange(len(hardware))
performance = np.array([18, 21, 64, 368, 530, 630, 863])

ax.barh(y_pos, performance, align='center')
ax.set_yticks(y_pos, labels=hardware)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Time of training per epoch in seconds (less = faster)')
ax.set_title('Training time of VGG16 NN 34 mln of parameters')

# plt.show()
plt.savefig('vgg16.png', bbox_inches='tight', pad_inches=1)
```
